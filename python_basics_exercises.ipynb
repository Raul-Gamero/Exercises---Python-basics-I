Print characters present at an even index number

```python
def even_index_chars(text):
    return text[::2]

# Example
print(even_index_chars("pynative"))
```

Check if first and last numbers of a list are the same

```python
def is_first_last_same(numbers):
    if len(numbers) == 0:
        return False
    return numbers[0] == numbers[-1]

#Example
print(is_first_last_same([10, 20, 30, 10]))
print(is_first_last_same([10, 20, 30, 40]))
```

Merge two lists (odd from first, even from second)

```python
def merge_lists(list1, list2):
    result = []

    for num in list1:
        if num % 2 != 0:
            result.append(num)

    for num in list2:
        if num % 2 == 0:
            result.append(num)

    return result

#Example
print(merge_lists([10, 20, 25, 30, 35], [40, 45, 60, 75, 90]))
```

Accept a list of 5 float numbers as input

```python
def get_float_list():
    floats = []
    for i in range(5):
        num = float(input(f"Enter float number {i+1}: "))
        floats.append(num)
    return floats

# Example (run manually)
numbers = get_float_list()
print(numbers)
```

Display numbers from a list using a loop

```python
def display_numbers(numbers):
    for num in numbers:
        print(num)

# Example
display_numbers([1, 2, 3, 4, 5])       
```

Print Full Multiplication Table

```python
def multiplication_table():
    for i in range(1, 11):
        for j in range(1, 11):
            print(i * j, end="\t")
        print()

# Example
multiplication_table() 
```

Create string from first, middle and last character

```python
def first_middle_last(text):
    first = text[0]
    middle = text[len(text)//2]
    last = text[-1]
    return first + middle + last


# Example
print(first_middle_last("pynative"))
```

Remove duplicates from list

```python
def remove_duplicates(items):
    return list(set(items))

# Example
print(remove_duplicates([1, 2, 3, 2, 4, 1, 5]))
```

```python
def remove_duplicates_ordered(items):
    unique = []
    for item in items:
        if item not in unique:
            unique.append(item)
    return unique
```

Delete a list of keys from a dictionary

```python
def delete_keys(dictionary, keys):
    for key in keys:
        dictionary.pop(key, None)
    return dictionary

# Example
sample_dict = {"name": "Kelly", "age": 25, "salary": 8000, "city": "New York"}
print(delete_keys(sample_dict, ["name", "salary"]))
```

Rename key of a dictionary

```python
def rename_key(dictionary, old_key, new_key):
    if old_key in dictionary:
        dictionary[new_key] = dictionary.pop(old_key)
    return dictionary

# Example
sample_dict = {"name": "Kelly", "age": 25, "salary": 8000}
print(rename_key(sample_dict, "salary", "income"))
    
```
